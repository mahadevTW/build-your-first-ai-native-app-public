## Cell 1 — Python version check

Must be 3.12 or newer.

```python
import sys

major = sys.version_info.major
minor = sys.version_info.minor

print(f"Python version : {sys.version}")
print(f"Kernel         : {sys.executable}")

if major == 3 and minor >= 12:
    print("\n✓  Python 3.12+ confirmed")
else:
    raise RuntimeError(
        f"\n✗  Python 3.{minor} found but 3.12+ is required.\n"
        "   Switch to the 'AI Meetup (Python 3.12)' kernel:\n"
        "   VS Code: click the kernel name in the top-right corner\n"
        "   Jupyter: Kernel menu → Change kernel"
    )
```

## Cell 2 — Package imports and versions

Verifies that `pip install -r requirements.txt` completed successfully.

```python
from importlib.metadata import version, PackageNotFoundError

required = [
    ("openai",       "openai"),
    ("pandas",       "pandas"),
    ("requests",     "requests"),
    ("jupyter",      "jupyter"),
    ("ipykernel",    "ipykernel"),
    ("notebook",     "notebook"),
    ("jupyterlab",   "jupyterlab"),
]

failures = []

print(f"{'Package':<16} {'Version':<15} Status")
print("-" * 44)

for import_name, package_name in required:
    try:
        __import__(import_name)
        ver = version(package_name)
        print(f"  {package_name:<14} {ver:<15} ✓")
    except (ImportError, PackageNotFoundError) as e:
        print(f"  {package_name:<14} {'—':<15} ✗  MISSING")
        failures.append(package_name)

if failures:
    raise ImportError(
        f"\n✗  Missing packages: {', '.join(failures)}\n"
        "   Fix: open a terminal, activate the venv, and run:\n"
        "     pip install -r requirements.txt"
    )

print("\n✓  All packages present")
```
